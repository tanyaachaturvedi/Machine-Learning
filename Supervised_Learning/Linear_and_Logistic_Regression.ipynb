```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# URL of the dataset
data_url = "http://lib.stat.cmu.edu/datasets/boston"

# Column names for the dataset
feature_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']

# Read the data, skipping initial rows and handling the irregular whitespace
df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)

# The data is in a strange format; features and target are interleaved
data = np.hstack([df.values[::2, :], df.values[1::2, :2]])
target = df.values[1::2, 2]

# Create a clean DataFrame for our features (X)
X = pd.DataFrame(data, columns=feature_names)

# Create a clean Series for our target (y)
y = pd.Series(target, name='MEDV')
```

```python
# 1. Create the new, combined DataFrame
dataset = X.copy()
dataset['MEDV'] = y

# 2. Check the first 5 rows (as you suggested)
# This lets you see the data and all the column names
print("--- .head() ---")
print(dataset.head())

# 3. Check for missing values and data types
# .info() is essential. It tells us if we have nulls and if our numbers
# are stored correctly (e.g., as 'float64' and not 'object').
print("\n--- .info() ---")
dataset.info()

# 4. Get a statistical summary of each column
# .describe() shows count, mean, standard deviation, min, max, etc.
# This is great for spotting outliers or understanding the data's scale.
print("\n--- .describe() ---")
print(dataset.describe())
```

# **Logistic Regression**


```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df=load_breast_cancer()
#independent features
X=pd.DataFrame(df['data'],columns=df['feature_names'])
```

```python
X.head()
```

```python
#dependent features
y=pd.DataFrame(df['target'],columns=['Target'])
y.head()
```

```python
y['Target'].value_counts()
```

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
params=[{'C':[1,5,10]}, {'max_iter': [100,150]}]
```

```python
log_reg_model=LogisticRegression(C=100,max_iter=100)
```

```python
model=GridSearchCV(log_reg_model,param_grid=params,scoring='f1',cv=5)
```

```python
model.fit(X_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
y_pred=model.predict(X_test)
y_pred
```

Confusion matrix

```python
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score
```

```python
confusion_matrix(y_test,y_pred)
```

```python
# Calculate the accuracy score
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
